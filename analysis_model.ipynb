```python
from sklearn.model_selection import train_test_split # splitting training and testing the data
from sklearn.preprocessing import StandardScaler  # scaling the data
from sklearn.linear_model import LogisticRegression # modelling side
from sklearn.metrics import classification_report, confusion_matrix # model evalation
from sklearn.pipeline import Pipeline # training the model and doing the transformation operations
from sklearn.compose import ColumnTransformer # data transformation
from sklearn.preprocessing import OneHotEncoder # data transformation
```

```python
df.head()
df_model=df.drop(["nameOrig","nameDest","isFlaggedFraud"],axis=1)
df_model.head()
```

```python
categorical=["type"]
numeric=["amount","oldbalanceOrg","newbalanceOrig","oldbalanceDest","newbalanceDest"]
y =df_model["isFraud"]
X=df_model.drop("isFraud",axis=1)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,stratify=y)
```

```python
preprocessor= ColumnTransformer(transformers=[("num",StandardScaler(),numeric),("cat",OneHotEncoder(drop="first"),categorical)],remainder="drop")
```

```python
pipeline=Pipeline([("prep",preprocessor),("clf" ,LogisticRegression( class_weight="balanced",max_iter=1000 ))     ])
```

```python
pipeline.fit(X_train,y_train)
```

```python
y_pred=pipeline.predict(X_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```

```python
pipeline.score(X_test,y_test) *100
```

```python
import joblib
joblib.dump(pipeline,"fraud_detection_pipeline.pkl")
```
